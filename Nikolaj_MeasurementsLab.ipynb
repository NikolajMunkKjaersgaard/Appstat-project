```python
def read_csv(filename):
    """Read CSV from Waveforms"""
    dat = np.genfromtxt(filename, delimiter=',', skip_header=13, names=True)
    time = dat['Time_s']
    voltage = dat['Channel_1_V']
    return time, voltage

def find_midpoints(time, voltage, show_plot=True):
    """Find timing of ball crossings"""
    Std_Voltage = np.std(voltage)
    Rolled_Voltage = np.roll(voltage, 1)
    Voltage_difference = np.roll(voltage - Rolled_Voltage, -1)
    Transient_Mask = np.abs(Voltage_difference) > Std_Voltage
    Transient_Times = np.array(time)[Transient_Mask]
    return Transient_Times

voltage = [1,2,3,4]
Rolled_Voltage = np.roll(voltage, 1)
Voltage_difference = np.roll(voltage - Rolled_Voltage, -1)
find_midpoints(voltage, voltage)
```
